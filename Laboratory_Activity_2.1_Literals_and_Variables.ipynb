**Procedure:**

1. A teacher wants to calculate the final grade in a CpE course and want to write it in a python program. The following are the requirements:

  I. PRELIM GRADE = 50% Prelim Exam + 50% Prelim Class Standing (CS)

  II. PRELIM CS = 50% Hands-on activity + 30% Quiz + 20% Assignment

  III. MIDTERM GRADE = 1/3 of PRELIM GRADE + 2/3 of (50% Midterm Exam + 50% Midterm Class Standing (CS))

  IV. MIDTERM CS = 50% Hands-on activity + 30% Quiz + 20% Assignment

  V. FINAL GRADE = 1/3 of MIDTERM GRADE + 2/3 of (50% Final Exam + 50% Final Class Standing (CS))

  VI. FINAL CS = 50% Hands-on activity + 30% Quiz + 20% Assignment
  
  VII. HOAs, Quizzes and Assignments are inputted as average of all submissions and are out of 100%.

  VIII. Major exams are inputted out of 100%.
  
  IX. Show the codes that successfully run the program.
  
  X. Provide comments or documentation strings for your program.

**Supplementary Activity**:

1. Test 3 students from the program you created.
2. The program should show the name of the student, the PRELIM, MIDTERM and FINAL grades.
3. Convert the final grade into the T.I.P.s numerical grade. Please refer to the grading system.



**Questions**: (write your answers in blue font color and questions in black)

**Conclusion**:

I was able to achieve the objectives of this activity by creating a Python program that uses literals and variables to compute the Prelim, Midterm, and Final grades correctly. I also used a loop to process multiple students, which made the program more efficient and helped me better understand how variables work within repeated executions.


```python
for i in range(1, 4):
    print(f"Student #{i}")
    stud = input("Enter your name: ")

    # PRELIM
    print("\nPRELIM")
    PHOA = float(input("Enter Hands-on Activity grade: "))
    PQUIZ = float(input("Enter Prelim Quiz grade: "))
    PAS = float(input("Enter Prelim Assignment grade: "))
    PCS = PHOA * 0.5 + PQUIZ * 0.3 + PAS * 0.2
    print("Prelim CS is:", "%.2f" % PCS)
    PEX = float(input("Enter Prelim Exam grade: "))
    PG = PEX * 0.5 + PCS * 0.5
    print("Prelim Grade is:", "%.2f" % PG)

    # MIDTERM
    print("\nMIDTERM")
    MHOA = float(input("Enter Hands-on Activity grade: "))
    MQUIZ = float(input("Enter Midterm Quiz grade: "))
    MAS = float(input("Enter Midterm Assignment grade: "))
    MCS = MHOA * 0.5 + MQUIZ * 0.3 + MAS * 0.2
    print("Midterm CS is:", "%.2f" % MCS)
    MEX = float(input("Enter Midterm Exam grade: "))
    MG = (1/3) * PG + (2/3) * (MEX * 0.5 + MCS * 0.5)
    print("Midterm Grade is:", "%.2f" % MG)

    # FINAL
    print("\nFINAL")
    FHOA = float(input("Enter Hands-on Activity grade: "))
    FQUIZ = float(input("Enter Final Quiz grade: "))
    FAS = float(input("Enter Final Assignment grade: "))
    FCS = FHOA * 0.5 + FQUIZ * 0.3 + FAS * 0.2
    print("Final CS is:", "%.2f" % FCS)
    FEX = float(input("Enter Final Exam grade: "))
    FG = (1/3) * MG + (2/3) * (FEX * 0.5 + FCS * 0.5)

    # Convert to T.I.P.S numerical grade
    if FG >= 97:
        FG = 1.0
    elif FG >= 94:
        FG = 1.25
    elif FG >= 91:
        FG = 1.5
    elif FG >= 88:
        FG = 1.75
    elif FG >= 85:
        FG = 2.0
    elif FG >= 82:
        FG = 2.25
    elif FG >= 79:
        FG = 2.5
    elif FG >= 76:
        FG = 2.75
    elif FG >= 75:
        FG = 3.0
    else:
        FG = 5.0

    print("Final Grade is:", "%.2f" % FG)
    print("\n")
```
