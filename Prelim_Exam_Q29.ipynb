Develop a python scripts that:

    Calculate total marks and percentage
    Assign grade for each subject based on marks:

90-100: A+

80-89: A

70-79: B+

60-69: B

50-59: C

40-49: D

Below 40: F (Fail)

    Determine overall result (Pass/Fail)



Performance Remarks:


Percentage ≥ 90: "Outstanding"

Percentage ≥ 80: "Excellent"

Percentage ≥ 70: "Very Good"

Percentage ≥ 60: "Good"

Percentage ≥ 40: "Satisfactory"

Below 40: "Needs Improvement"


```python
#PRELIM EXAM Q29
def grade(mark):
    if mark >= 90:
        return "A+"
    elif mark >= 80:
        return "A"
    elif mark >= 70:
        return "B+"
    elif mark >= 60:
        return "B"
    elif mark >= 50:
        return "C"
    elif mark >= 40:
        return "D"
    else:
        return "F (FAIL)"

def remark(percentage):
    if percentage >= 90:
        return "Outstanding"
    elif percentage >= 80:
        return "Excellent"
    elif percentage >= 70:
        return "Very Good"
    elif percentage >= 60:
        return "Good"
    elif percentage >= 40:
        return "Satisfactory"
    else:
        return "Needs Improvement"

#User Input
subjects = []
marks = []

n = int(input("Enter number of subjects: "))

for i in range(n):
    subject = input(f"Enter subject {i+1} name: ")
    mark = int(input(f"Enter marks for {subject}: "))
    subjects.append(subject)
    marks.append(mark)

total = sum(marks)
percentage = total / n

print("\nSubject Grades:")
for i in range(n):
    print(subjects[i], ":", marks[i], "->", grade(marks[i]))

#Result
if all(m >= 40 for m in marks):
    result = "Pass"
else:
    result = "Fail"

print("\nTotal Marks:", total)
print("Percentage:", round(percentage, 2), "%")
print("Overall Result:", result)
print("Performance Remark:", remark(percentage))
```
