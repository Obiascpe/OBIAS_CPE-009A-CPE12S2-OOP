Assignment Operators

```python
#" = "
x = 5
print(x)

#" += "
s = 1
s += 3
print(s)

#" -= "
s -= 3
print(s)

#" *= "
s *= 3
print(s)

#" /= "
s /= 3
print(s)

#" %= "
s %= 3
print(s)

#" //= "
s //= 3
print(s)

#" **= "
s **= 3
print(s)

#" &= "
s = 1
s &= 3
print(s)

#" |= "
s |= 3
print(s)

#" ^= "
s ^= 3
print(s)

#" >>= "
s >>= 3
print(s)

#" <<= "
s <<= 3
print(s)
```

Comparison Operators

```python
#Given:
t, u = 1, 2

#Equal "x == y"
print(t == u)

#Not Equal "x != "
print(t != u)

#Greater than "x > y"
print(t > u)

#Less than "x < y"
print(t < u)

#Greater than or Equal to "x >= y"
print(t >= u)

#Less than or Equal to "x <= y"
print(t <= u)
```

Logical Operators

```python
#Given:
t, u = 1, 2

#And "x < 5 and x < 10"
print(t < 5 and u < 10)

#Or "x < 5 or x < 10"
print(t < 5 or u < 10)

#Not "not(x < 5 and x < 10)"
print(not(t < 5 and u < 10))
```

Identity Operators

```python
#Given:
t, u = 1, 2

#Is "x is y"
print(t is u)

#Is not "x is not y"
print(u is not t)
```

Bitwise Operators

```python
#Given:
v = 2

#AND(&)
v = 2
print(v & 1)

#OR(|)
v = 2
print(v | 1)


#XOR(^)
v = 2
print(v ^ 1)

#NOT(~)
v = 2
print(~v)

#Zero fill left shift(<<)
v = 2
print(v << 1)

#Signed right shift(>>)
v = 2
print(v >> 1)
```
