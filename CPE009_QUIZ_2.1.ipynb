```python
def termCrit(actNumber):
    rawScore = []
    totalScore = []

    for i in range(actNumber):
        inputScore = float(input(f"\nScore #{i+1}: "))
        inputTotal = float(input(f"Total #{i+1}: "))

        if inputScore > inputTotal or inputTotal == 0:
            print("\nInvalid Scores.\n")
            continue
        else:
            rawScore.append(inputScore)
            totalScore.append(inputTotal)

    if sum(totalScore) == 0:
        return 0

    grade = (sum(rawScore) / sum(totalScore)) * 100
    print(f"\nAverage Grade: {grade:.2f}%")
    return grade


result = termCrit(int(input("How many hands-on activities this term?\nHOA: ")))
```

```python
def termCS():
  termHOA = termCrit(int(input("How many hands-on activities this term?\nHands-on Activities: ")))

  termQuiz = termCrit(int(input("\nHow many quizzes this term?\nQuizzes: ")))

  termAss = termCrit(int(input("\nHow many assignments?\nAssignments: ")))

  grade = (0.5 * termHOA) + (0.3 * termQuiz) + (0.2 * termAss)
  return grade

print(f"\nClass Standing: {termCS():.2f}%")
```

```python
def termGrade():
  CSgrade = termCS()

  print("\nComputing for the term grade.")
  examRaw = float(input("Exam Score: "))
  examTotal = float(input("Exam Total: "))
  examGrade = (examRaw/examTotal) * 100

  grade = (0.5 * examGrade) + (0.5 * CSgrade)
  return grade

print(f"\nTermGrade: {termGrade():.2f}%")
```

```python
def semesterGrade():
  print("PRELIM GRADE")
  prelimGrade = termGrade()

  print("\nMIDTERM GRADE")
  midtermGrade = termGrade()
  finalMG = (1/3)*prelimGrade + (2/3)*midtermGrade

  print("\nFINAL GRADE")
  finalGrade = termGrade()
  finalFG = (1/3)*finalMG + (2/3)*finalGrade

  return finalFG

print(f"\nSemester Grade: {semesterGrade():.2f}%")
```
